# Importing important libraries and modules

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.model_selection import KFold,cross_val_score,GridSearchCV
import numpy as np
```

```python
from sklearn.datasets import load_iris
dataset=load_iris()
dataset
```

```python
df=pd.DataFrame(dataset.data,columns=dataset.feature_names)
df.head()
```

```python
df['Target']=dataset.target
```

```python
df.head()
```

```python
x=df.drop('Target',axis=1)
y=df['Target']
```

```python
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier,RandomForestClassifier
from sklearn.linear_model import LogisticRegression
```

```python
kernels=['linear','rbf']
c_params=[10,20,50]
avg_score={}
for k in kernels:
    for c in c_params:
        cv_score=cross_val_score(SVC(kernel=k,C=c,gamma='auto'),dataset.data,dataset.target,cv=5)
        avg_score[f'{k}_{c}']=np.average(cv_score)
avg_score
```

```python
param_grid={'kernel':['linear','rbf'],
            'C':[10,20,50]}
```

```python
#doing same operation which is done by above code using gridserch cv
clf=GridSearchCV(SVC(gamma='auto'),param_grid,return_train_score=False,cv=5)
```

```python
clf.fit(x,y)
```

```python
clf.cv_results_
```

```python
table_df=pd.DataFrame(clf.cv_results_)
table_df
```
