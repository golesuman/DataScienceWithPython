```python
import pandas as pd
from sklearn import linear_model
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_excel('drop.xlsx')
df
```

```python
dumdf=pd.get_dummies(df.city)
dumdf
```

```python
mergeddf=pd.concat([df,dropdf], axis='columns')
mergeddf
```

```python
dfdrop=mergeddf.drop(['city','lalitpur'],axis='columns')
dfdrop
```

```python
X=dfdrop.drop(['price'],axis='columns')
X
```

```python
Y=dfdrop.price
Y
```

```python
pre=linear_model.LinearRegression()
pre.fit(X,Y)
```

```python
pre.coef_
```

```python
pre.intercept_
```
