```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
foodpr={"Food":["Samosa","Momo", "Chowmein",
"Pau"],
"price":[20,80,70,30]
}
df=pd.DataFrame(foodpr)
df
```

```python
plt.bar(df.Food,df.price)
```

```python
plt.axis("equal")
plt.pie(df.price,labels=df.Food,radius=3, autopct="%0.2f%%", explode=[0.2,0,0,0.2])
plt.show()
```
